```python
# 판다스 모듈 불러오기
import pandas as pd

# 엑셀 파일 불러오기(엑셀 파일명 넣으면 됨.)
df_origin = pd.read_excel('SNP_for_python.xlsx')
df_origin

# 컬럼마다 값을 추출해서 리스트에 각각 저장하기
list_location = df_origin['location'].values.tolist()
list_note1 = df_origin['note1'].values.tolist()
list_note2 = df_origin['note2'].values.tolist()
list_label1 = df_origin['label1'].values.tolist()
list_label2 = df_origin['label2'].values.tolist()

# 값에 포함되어 있는 불필요한 문자열 제거하기
list_label1 = [label1.replace("\t", "") for label1 in list_label1]
list_label2 = [label2.replace("\t", "").replace(".","") for label2 in list_label2]

# 반복문 이용해서 형식에 맞게 값 출력하기
for i in range(959):
    print(f"""
     variation       {list_location[i]}
                     /note={list_note1[i]}
                     /note={list_note2[i]}
                     /label={list_label1[i]}/{list_label2[i]}""", end='')
    
```
